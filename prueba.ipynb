```python
vendedores = ['Ana', 'Luis', 'Marta', 'Sergio']
Ventas = [1500.50, 2300.75, 1800.00, 2100.25]

#vendedores.append('Camila')

meses = []
ventas = []


cantidad_meses = int(input('Ingrese el número de meses a registrar: '))

for i in range(cantidad_meses):
    
    mes = input('Ingrese el nombre del mes: ')
    venta = float(input(f'Ingrese el total de ventas en {mes}: '))

    # Guardar los datos
    meses.append(mes)
    ventas.append(venta)
    
n = len(meses)

print('\nRegistro de Ventas Mensuales:')
for i in range(n):
    print(f'{meses[i]}: ${ventas[i]:,.1f}')
```

```python
temperaturas = ["10 C", "15 C", "20 C", "25 C", "30 C", "35 C", "40 C"]

for i in range(len(temperaturas)):
    temperaturas[i] = int(temperaturas[i].strip(" C"))

temperaturas_filtradas = []

temperaturas_filtradas.append(temperaturas)
temperaturas_frias = []
temperaturas_calidas = []

print(temperaturas_filtradas)
for temp in temperaturas:
    if temp <= 18:
        temperaturas_frias.append(temp)
    else:
        temperaturas_calidas.append(temp)

#print(temp)

print("Temperaturas frías:", temperaturas_frias)
print("Temperaturas cálidas:", temperaturas_calidas)
```
